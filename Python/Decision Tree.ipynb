```python
pp = {
    "checking_balance" : {"unknown":0,"< 0 DM":1, "1 - 200 DM":2,"> 200 DM":3},
    "credit_history" : {"critical": 0,"poor":1,"good":2,"very good":3,"perfect":4},
    "savings_balance" : {"unknown":0,"< 100 DM":1, "100 - 500 DM":2,"500 - 1000 DM":3,"> 1000 DM":4},
    "employment_duration":{"unemployed":0,"< 1 year":1,"1 - 4 years":2,"> 7 years":4,"4 - 7 years":3,},
    "phone":{"no":0,"yes":1},
    "default":{"no":0,"yes":1}
    }
obj = ["purpose","housing","other_credit","job"]
```

```python

df.select_dtypes(include = 'object')
```

```python
df = df.replace(pp)
df = pd.get_dummies(df,columns =obj)
```

```python
df.head()
```

```python
df.info()
```

```python
x = df.drop(columns = 'default',axis = 1)
y = df["default"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 23) 
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(criterion ='gini',random_state = 23)
dt_model.fit(X_train,y_train)
```

```python
print(dt_model.score(X_train,y_train))
print(dt_model.score(X_test,y_test))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.metrics import roc_auc_score
```

```python
md = [3,5,10,25]
ml = [15,30 ,70]
ms = [15,30,70]
model_performance_data = []

for i in md:
    for j in ml:
        for k in ms:
            
            mp = DecisionTreeClassifier(criterion = 'gini',max_depth = i, min_samples_leaf = j,random_state = 23)
            dt_model = mp.fit(X_train,y_train)
            test_pred = dt_model.predict_proba(X_test)
            train_pred = dt_model.predict_proba(X_train)
            
            t1 = "max_depth" + str(i)+"_min_leaf"+str(j)+"_min_spit"+str(k)
            t2 = [t1,round(roc_auc_score(y_train,train_pred[:,1]),4),round(roc_auc_score(y_test,test_pred[:,1]),4)]
            
            model_performance_data.append(t2)
```

# Post Puring

```python
apl = [0.001,0.005,0.01,0.05,0.1]
for k in apl:
    
    mp =DecisionTreeClassifier(criterion ='gini',ccp_alpha = k,random_state = 23)
    dt_model = mp.fit(X_train,y_train)
    test_pred = dt_model.predict_proba(X_test)
    train_pred = dt_model.predict_proba(X_train)
    
    t1 = "ccp_alpha"+str(k)
    t2 = [t1,round(roc_auc_score(y_train,train_pred[:,1]),4),round(roc_auc_score(y_test,test_pred[:,1]),4)]
    
    model_performance_data.append(t2)
```

```python
model_performance_df = pd.DataFrame(model_performance_data)
model_performance_df.rename(columns ={0 :'DESC',1:'Train_AUC',2:'Test_AUC'},inplace = True)
```

```python
model_performance_df.sort_values(by = ['Test_AUC','Train_AUC'],ascending = [False,False])
```

# Second round of Tunning

```python
md = [5,7,9]
ml = [15,20,25]
apl = [0.003,0.004,0.005,0.006,0.007]

model_performance_data =[]

for i in md:
    for j in ml:
        for k in apl:
            
            mp = DecisionTreeClassifier(criterion = 'gini',max_depth = i, min_samples_leaf = j,random_state = 23)
            dt_model = mp.fit(X_train,y_train)
            test_pred = dt_model.predict_proba(X_test)
            train_pred = dt_model.predict_proba(X_train)
            
            t1 = "max_depth" + str(i)+"_min_leaf"+str(j)+"_min_spit"+str(k)
            t2 = [t1,round(roc_auc_score(y_train,train_pred[:,1]),4),round(roc_auc_score(y_test,test_pred[:,1]),4)]
            
            model_performance_data.append(t2)
```

```python
model_performance_df = pd.DataFrame(model_performance_data)
model_performance_df.rename(columns ={0 :'DESC',1:'Train_AUC',2:'Test_AUC'},inplace = True)
```

```python
model_performance_df.sort_values(by = ['Test_AUC','Train_AUC'],ascending = [False,False])
```

# Grid Search

```python
from sklearn.model_selection import GridSearchCV

grid = {'ccp_alpha' : [0.002, 0.003, 0.004],
           'max_depth' : [7,8,9],
           'min_samples_leaf':[15,18,20,22,25]
       }

dt_model = DecisionTreeClassifier(random_state = 23)
grid_search = GridSearchCV(estimator = dt_model, param_grid = grid,cv = 5, verbose = True)
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_estimator_
```

# Final Model

```python
mp = DecisionTreeClassifier(criterion = 'gini',max_depth = 7, min_samples_leaf = 18, ccp_alpha = 0.003,random_state = 23)
dt_model = mp.fit(X_train,y_train)

test_pred = dt_model.predict_proba(X_test)
train_pred = dt_model.predict_proba(X_train)

print("Performance on training data :",round(roc_auc_score(y_train,train_pred[:,1]),4))
print("performance on testing data :",round(roc_auc_score(y_test,test_pred[:,1]),4))
```
